```python
import pandas as pd
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2024_per_game.html"
html = pd.read_html(url, header = 0)
df = html[0]
```

```python
raw = df.drop(df[df.Age == "Age"].index)
raw = raw.fillna(0)
playerstats = raw.drop(["Rk"], axis=1)
```
